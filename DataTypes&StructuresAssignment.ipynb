```python
# 1. What are data structures, and why are they important?

'''
A data structure is a logical or mathematical model for organizing data in memory, determining how data values are arranged and related.
They provide a blueprint for how data is stored, accessed, and manipulated (e.g., inserted, deleted, searched).
Examples: Arrays, Linked Lists, Stacks, Queues, Trees, Graphs, Hash Tables
They are important due to following reasons:
1. Efficiency: They allow for faster operations (searching, sorting, inserting) by arranging data optimally, reducing time and memory usage (space complexity).
2. Scalability: Essential for managing large, complex datasets efficiently in databases, operating systems, and internet services.
3. Foundation for Algorithms: They are the core components that algorithms work on; the choice of data structure heavily impacts an algorithm's efficiency.
4. Code Organization: They simplify code by providing structured ways to handle information, making programs easier to understand and maintain.
5. Problem Solving: Different structures solve different problems; choosing the right one (e.g., a Queue for task scheduling vs. a Tree for hierarchical data) is key to good software design.
'''
```

```python
# 2. Explain the difference between mutable and immutable data types with examples.

'''
Mutable Data Types:
A mutable object or data type can be modified in place. When you change the value of a mutable object, its memory address remains the same.
The data within the existing object is altered directly.
Examples: Lists, Dictionaries, Sets

Immutable Data Types:
An immutable object or data type cannot be changed after it is created. Any operation that appears to modify an immutable object actually
creates a new object in memory with the updated data. The original object remains unchanged.
Examples: Integers, Floats, Strings, Tuples
'''
```

```python
# 4. Describe how dictionaries store data.

'''
Dictionaries (also known as hash maps or associative arrays in computer science) primarily store data
using a hash table structure, which allows for fast data retrieval using unique keys rather than sequential indexes.

Data is stored as a collection of key-value pairs, where each unique key is associated with a specific value.

1. Keys: Act as unique identifiers and must be of an immutable data type (like strings, numbers, or tuples) so they can be "hashed" consistently.

2. Values: The actual data associated with a key. Values can be of any data type, including other dictionaries or lists, and do not need to be unique.
'''
```

```python
# 5. Why might you use a set instead of a list in Python.

'''
Key reasons for choosing a set include:

1. Fast Membership Testing: Checking if an element exists within a set (e.g., using the in operator) is significantly faster,
   on average O(1) time complexity, compared to a list which has O(n) complexity.

2. Automatic Uniqueness: A set inherently only stores unique elements. Duplicate additions are ignored automatically.
   If you need a collection of items where each item appears only once, a set simplifies the code.

3. Mathematical Operations: set objects support mathematical set operations like union, intersection, and difference,
   which can be very efficient for tasks involving comparisons between collections.
'''
```

```python
# 14. How does the “in” keyword work differently for lists and dictionaries?

'''
The in keyword performs a membership test, but its underlying mechanism and speed differ drastically
between lists and dictionaries.

1. Mechanism: Linear vs. Hashed Search
   A. Lists (Linear Search): To check if a value exists in a list, Python starts at the first element
      and checks each one sequentially until it finds a match or reaches the end.
   B. Dictionaries (Hashed Search): Python uses a hash table for dictionaries. It runs the "key" through
      a hash function to get a specific memory address and jumps directly to that spot to see if anything is there.

2. Performance (Big O Complexity): As your data grows, the difference in speed becomes massive

   Data Structure,              Average Time Complexity,            Behavior with Large Data
   List,                        O(n),                               Slows down. Doubling the list size doubles the search time.
   Dictionary,                  O(1),                               Constant speed. Searching a dictionary with 10 items takes nearly the same time as one with 10 million items.

3. Search Target: Value vs. Key
   A. List: item in my_list searches for the actual value within the list.
   B. Dictionary: item in my_dict only searches for keys by default. It does not check the values.
      To check if a value exists in a dictionary, you must use item in my_dict.values(), but this reverts the performance to a slow O(n) linear search.
'''

my_list = [1, 2, 3]
my_dict = {1: "a", 2: "b", 3: "c"}

# List: Searches for the number 2 in the elements
print(2 in my_list)  # True

# Dictionary: Searches for the number 2 in the KEYS
print(2 in my_dict)  # True

# Dictionary: Does NOT search values by default
print("b" in my_dict) # False (checks keys, not values)
```

```python
# 15. Can you modify the elements of a tuple? Explain why or why not?

'''
No, you cannot modify the elements of a tuple once it has been created.

Tuples are immutable by design. In Python, this means the memory block allocated for the tuple is fixed.
Unlike a list, a tuple does not have "slots" that can be reassigned to new objects.
If you attempt to change an element (e.g., my_tuple[0] = 5), Python will raise a TypeError.

The "Exception": Mutable Objects Inside a Tuple
While the tuple itself cannot be changed, it can contain mutable objects (like a list). In this case:
You cannot replace the list object with a different object.
You can modify the contents of that list.
'''

my_tuple = (1, 2, [3, 4])
# my_tuple[2] = [5, 6]  <-- Error: Cannot change the tuple's reference
my_tuple[2][0] = 99     # Works: Modifying the list INSIDE the tuple
print(my_tuple)         # Output: (1, 2, [99, 4])
```

```python
# 16. What is a nested dictionary, and give an example of its use case?

'''
A nested dictionary is a dictionary that contains other dictionaries as its values.
This creates a hierarchical structure, allowing you to represent complex, related data in a tree-like format.

Use Case: Structured Records
The most common use case is storing data for multiple entities that share the same attributes,
such as a user database, a product catalog, or a nested API response.

Example: Employee Database
Instead of having separate lists for names, roles, and emails, you can group all
information under a unique identifier (like an Employee ID).
'''

employees = {
    "emp_01": {
        "name": "Alice",
        "position": "Developer",
        "skills": ["Python", "SQL"]
    },
    "emp_02": {
        "name": "Bob",
        "position": "Designer",
        "skills": ["Figma", "CSS"]
    }
}

# Accessing data:
print(employees["emp_01"]["name"])    # Output: Alice
print(employees["emp_02"]["skills"])  # Output: ['Figma', 'CSS']
```

```python
# 17. Describe the time complexity of accessing elements in a dictionary.

'''
The time complexity of accessing elements in a Python dictionary is O(1) on average.
This is known as constant time, meaning the speed of access remains the same regardless
of whether the dictionary contains 10 items or 10 million items.

How it achieves O(1)? Dictionaries use a data structure called a hash table.
Instead of searching through a list from top to bottom, Python follows these steps:

1. Hashing: It takes your key and runs it through a hash function to generate a unique integer.

2. Indexing: It uses that integer to calculate a specific index (position) in an underlying array.

3. Direct Access: It jumps directly to that memory address to retrieve the value.

Operation,                       Dictionary,             List
Access by Key/Index,             O(1),                   O(1)
Search by Value,                 O(n),                   O(n)
Membership Check (in),           O(1),                   O(n)
'''
```

```python
# 19. Why are dictionaries considered unordered, and how does that affect data retrieval?

'''
Dictionaries are built using Hash Tables. In a hash table, the location of an item in memory
is determined by a hash function applied to the key, not by its sequence.

How it Affects Data Retrieval?
Because dictionaries are mappings rather than sequences, data retrieval behaves differently than in a list:

1. No Positional Indexing: You cannot retrieve the "third item" using my_dict[2]. This will raise a
   KeyError unless 2 is an actual key. You must know the specific label (key) to get the data.

2. Constant Search Speed: Because retrieval uses a hash to jump directly to a memory address,
   finding an item takes the same amount of time (O(1)) whether it was the first item added or the millionth.

3. Slicing is Not Possible: You cannot "slice" a dictionary (e.g., my_dict[0:5])
   because there is no numerical sequence to slice across.
'''
```

```python
# 20. Explain the difference between a list and a dictionary in terms of data retrieval.

'''
The fundamental difference in data retrieval between a list and a dictionary lies in
how you find the data and how long it takes as the collection grows.

1. Retrieval Method: Index vs. Key
   A. List (Sequential): You retrieve data based on its position (index). You must know where the item is located numerically (e.g., the 0th, 1st, or 2nd element).
   B. Dictionary (Mapping): You retrieve data based on a unique label (key). You don't need to know where the data is stored; you just need to know its name.

2. Retrieval Speed (Time Complexity): This is the most significant technical difference.
   Feature,                             List Retrieval,                                                                    Dictionary Retrieval
   By Index/Key,                        Fast (O(1)): Jumping to a known index is instant.,                                 Fast (O(1)): Hashing the key leads directly to the data.
   Searching by Value,                  Slow (O(n)): Python must check every item one by one from the start.,              Slow (O(n)): Searching by value (not key) requires scanning the whole dict.
   Membership (in),                     Slow (O(n)): Scanning the list until the item is found.,                           Fast (O(1)): Hashing the key tells Python instantly if it exists.
'''
```
